```python
# Extraction and Checking data

import os

alpha_folders = sorted([
    f for f in os.listdir(SignAlphaSet_Path)
    if os.path.isdir(os.path.join(SignAlphaSet_Path, f))
])

print(f"Found {len(alpha_folders)} classes: {alpha_folders}")

for folder in alpha_folders:
    folder_path = os.path.join(SignAlphaSet_Path, folder)
    num_images = len([
        f for f in os.listdir(folder_path)
        if f.lower().endswith(('.jpg', '.png', '.jpeg'))
    ])
    print(f"{folder}: {num_images} images")
```

```python
# Final check for mounting

from PIL import Image
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 2, figsize=(10, 6))
axes = axes.flatten()

for i, folder in enumerate(alpha_folders[:4]):
    folder_path = os.path.join(SignAlphaSet_Path, folder)
    img_file = os.listdir(folder_path)[0]
    img = Image.open(os.path.join(folder_path, img_file))

    axes[i].imshow(img)
    axes[i].set_title(f"Letter: {folder}")
    axes[i].axis('off')

plt.tight_layout()
plt.savefig('samples.png')
plt.show()

print("\nSaved sample images to 'samples.png'")
```
